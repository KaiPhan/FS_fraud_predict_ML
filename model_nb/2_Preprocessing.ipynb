```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_excel('DataFinal_v2.xlsx')
df = df.drop(['Unnamed: 0','Count', 'CountBS', 'CountIS', 'CountCF'], axis = 1)
df
```

```python
#Z-Score

#Z Score = 3.25 + 6.56A + 3.26B + 6.72C + 1.05D 
#Trong đó:
#A = Vốn lưu động / tổng tài sản
##Vốn lưu động = Tài sản ngắn hạn – Nợ ngắn hạn
#A = Working Capital (Current Assets – Current Liabilities) / Total Assets (Measures liquidity of firm)
df["Z_A"] = ((df["I. TÀI SẢN NGẮN HẠN Hợp nhất"] - df["1. Nợ ngắn hạn Hợp nhất"]) 
               / df["A. TỔNG CỘNG TÀI SẢN Hợp nhất"])
#B = Thu nhập giữ lại / tổng tài sản
## Lợi nhuận giữ lại = Lợi nhuận giữ lại ban đầu (số còn lại từ các năm trước đó) + Thu nhập ròng – cổ tức
### Thu nhập ròng = Tổng Doanh thu – Tổng chi phí
#B= Retained Earnings / Total Assets (measures accumulated profits compared to assets)
df["Z_B"] = (df["1.12.2. LNST chưa phân phối kỳ này Hợp nhất"]/ df["A. TỔNG CỘNG TÀI SẢN Hợp nhất"])

#C = Thu nhập trước lãi và thuế (EBIT) / tổng tài sản
##EBIT = Tổng lợi nhuận kế toán trước thuế + Chi phí lãi vay
#C= Earnings Before Interest & Taxes (EBIT) / Total Assets (measures how much profit the firms assets are producing)
df["Z_C"] = ((df['16. Tổng lợi nhuận kế toán trước thuế Hợp nhất'] 
                + df['7.1. Trong đó: Chi phí lãi vay Hợp nhất']) / df["A. TỔNG CỘNG TÀI SẢN Hợp nhất"])

#D = Equity/ Total liabilities
df["Z_D"] = df['II. VỐN CHỦ SỞ HỮU Hợp nhất']  / df['I. NỢ PHẢI TRẢ Hợp nhất']


df["zscore"] = 3.25 + 6.56 * df["Z_A"] + 3.26 * df["Z_B"] + 6.72 * df["Z_C"] + 1.05 * df["Z_D"] 
                      
##Table:
df_Z = df[["Tên","Z_A","Z_B","Z_C","Z_D","zscore"]]
df_Z
```

```python
    # Create prvious year data
df['sale'] = df['3. Doanh thu thuần Hợp nhất']
df['pre_sale'] = df.sort_values(['Year']).groupby('Mã CK')['3. Doanh thu thuần Hợp nhất'].shift()
df['cogs'] = df['4. Giá vốn hàng bán Hợp nhất']
df['pre_cogs'] = df.sort_values(['Year']).groupby('Mã CK')['4. Giá vốn hàng bán Hợp nhất'].shift()
df['rect'] = df['1. Phải thu dài hạn Hợp nhất'] + df['3. Các khoản phải thu ngắn hạn Hợp nhất']
df['pre_rect'] = df.sort_values(['Year']).groupby('Mã CK')['rect'].shift()
df['act'] = df['I. TÀI SẢN NGẮN HẠN Hợp nhất']
df['pre_act'] = df.sort_values(['Year']).groupby('Mã CK')['act'].shift()
df['ppegt'] = df['2.1. GTCL TSCĐ hữu hình Hợp nhất']
df['pre_ppegt'] = df.sort_values(['Year']).groupby('Mã CK')['ppegt'].shift()
df['dp'] = df['2.1.2. Khấu hao lũy kế TSCĐ hữu hình Hợp nhất'] + df['2.2.2. Khấu hao lũy kế tài sản thuê tài chính Hợp nhất'] + df['2.3.2. Khấu khao lũy kế TSCĐ vô hình Hợp nhất']
df['pre_dp'] = df.sort_values(['Year']).groupby('Mã CK')['dp'].shift()
df['at'] = df['A. TỔNG CỘNG TÀI SẢN Hợp nhất']
df['pre_at'] = df.sort_values(['Year']).groupby('Mã CK')['at'].shift()
df['xsga'] = df['9. Chi phí bán hàng Hợp nhất'] + df['10. Chi phí quản lý doanh  nghiệp Hợp nhất']
df['pre_xsga'] = df.sort_values(['Year']).groupby('Mã CK')['xsga'].shift()
df['ni'] = df['11. Lợi nhuận thuần từ hoạt động kinh doanh Hợp nhất']
df['pre_ni'] = df.sort_values(['Year']).groupby('Mã CK')['ni'].shift()
df['oancf'] = df['I. Lưu chuyển tiền thuần từ hoạt động kinh doanh (GT) Hợp nhất']
df['pre_oancf'] = df.sort_values(['Year']).groupby('Mã CK')['oancf'].shift()
df['lct'] = df['1. Nợ ngắn hạn Hợp nhất']
df['pre_lct'] = df.sort_values(['Year']).groupby('Mã CK')['lct'].shift()
df['dltt'] = df['2. Nợ dài hạn Hợp nhất']
df['pre_dltt'] = df.sort_values(['Year']).groupby('Mã CK')['dltt'].shift()
df['sec'] = df['2.3. Đầu tư nắm giữ đến ngày đáo hạn Hợp nhất'] + df['5.5. Đầu tư nắm giữ đến ngày đáo hạn Hợp nhất']
df['pre_sec'] = df.sort_values(['Year']).groupby('Mã CK')['sec'].shift()
    
    # Create derived variable
df['asset_qual'] = (1 - (df['act'] + df['ppegt'] + df['sec']))/df['at']
df['pre_asset_qual'] = (1 - (df['pre_act'] + df['pre_ppegt'] + df['pre_sec']))/df['pre_at']
   
df['DSRI'] = (df['rect']/df['sale']) / (df['pre_rect'] / df['pre_sale'])
df['GMI'] =  ((df['pre_sale']-df['pre_cogs'])/df['pre_sale']) / ((df['sale']-df['cogs'])/df['sale'])
df['AQI'] = df['asset_qual'] / df['pre_asset_qual']
df['SGI'] = df['sale'] / df['pre_sale']
df['DEPI'] = (df['pre_dp'] / (df['pre_dp'] + df['pre_ppegt'])) / (df['dp'] / (df['dp'] + df['ppegt']))
df['SGAI'] = (df['xsga'] / df['sale']) / (df['pre_xsga'] / df['sale'])
df['TA'] = (df['ni'] - df['oancf']) / df['at']
df['LVGI'] = ((df['lct'] + df['dltt']) / df['at']) / ((df['pre_lct'] + df['pre_dltt']) / df['pre_at'])
    
    # M-score
df['mscore'] = -4.84 + .920 * df['DSRI'] + .528 * df['GMI'] + .404 * df['AQI'] \
                    + .892 * df['SGI'] + .115 * df['DEPI'] - .172 * df['SGAI'] + 4.679 * df['TA'] - .327 * df['LVGI']
```
