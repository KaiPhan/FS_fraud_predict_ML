# Combining raw data

```python
df2010 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2010')
df2010 = df2010.drop('STT', axis = 1)
df2010['Year'] = '2010'

df2011 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2011')
df2011 = df2011.drop('STT', axis = 1)
df2011['Year'] = '2011'

df2012 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2012')
df2012 = df2012.drop('STT', axis = 1)
df2012['Year'] = '2012'

df2013 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2013')
df2013 = df2013.drop('STT', axis = 1)
df2013['Year'] = '2013'

df2014 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2014')
df2014 = df2014.drop('STT', axis = 1)
df2014['Year'] = '2014'

df2015 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2015')
df2015 = df2015.drop('STT', axis = 1)
df2015['Year'] = '2015'

df2016 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2016')
df2016 = df2016.drop('STT', axis = 1)
df2016['Year'] = '2016'

df2017 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2017')
df2017 = df2017.drop('STT', axis = 1)
df2017['Year'] = '2017'

df2018 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2018')
df2018 = df2018.drop('STT', axis = 1)
df2018['Year'] = '2018'

df2019 = pd.read_excel('DataFull2_2.xlsx', sheet_name='2019')
df2019 = df2019.drop('STT', axis = 1)
df2019['Year'] = '2019'
```

```python
frames = [df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019]

df_final = pd.concat(frames)
df_final
```
